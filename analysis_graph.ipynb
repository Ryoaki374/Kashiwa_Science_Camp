```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./graph_preset.mplstyle')
```

```python
npz_data =  np.load('./data/mockdata_sample_p.npz',)
for key in npz_data.files:
    print(f"  {key:<8}: (rows, columns) = {npz_data[key].shape}")

data = dict(npz_data)
#data['data'] = 10**(data['data']/10)
```

```python
fig,ax = plt.subplots(figsize=(8,6))
ax.scatter(data['freq'],data['trans'],label='p-pol',color='r',s=8)
ax.set_ylim(-0.05,1.1)
ax.set_xlim(36,161)
ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Transmittance')
ax.legend(loc='lower right')
```
