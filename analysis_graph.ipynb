```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./graph_preset.mplstyle')
```

```python
data =  np.loadtxt('data/mockdata_sample.csv', delimiter=',', dtype='float',skiprows=1)

data[:,1]  = 10**(data[:,1]/10)
data[:,2]  = 10**(data[:,2]/10)
```

```python
fig,ax = plt.subplots(figsize=(8,6))
ax.scatter(data[:,0],data[:,1],label='p-pol',color='r',s=8)
ax.scatter(data[:,0],data[:,2],label='s-pol',color='b',s=8)
#ax.set_ylim(0,1.1)
ax.set_xlim(34,161)
ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Transmittance')
ax.legend(loc='lower right')
```
